```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("diabetesIndo.csv")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # Last column

# Plot scatter plots for all features against the target variable
plt.figure(figsize=(12, 10))
for i, column in enumerate(X.columns):
    plt.subplot(3, 3, i+1)
    plt.scatter(X[column], y, alpha=0.7)
    plt.title('Scatter Plot of {} vs Luaran'.format(column))
    plt.xlabel(column)
    plt.ylabel('Luaran')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
