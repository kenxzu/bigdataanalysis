**SQL LIte in Python**

SQLite is a lightweight, serverless, self-contained, and transactional SQL database engine that is widely used due to its simplicity, reliability, and ease of use. In Python, SQLite is supported through the built-in sqlite3 module, which allows you to interact with SQLite databases directly from your Python code.

**Using SQLite in Python**


In Python, SQLite databases can be accessed using the sqlite3 module, which comes pre-installed with Python.
You can connect to an SQLite database using the connect() function, providing the filename of the database file as an argument.
Once connected, you can execute SQL queries using the execute() method of the connection object.
SQLite supports standard SQL syntax, allowing you to perform various operations such as creating tables, inserting data, querying data, updating records, and deleting records.

**Creating Tables**

You can create tables in an SQLite database using the CREATE TABLE SQL statement.
Define the columns of the table along with their data types and constraints.
Example: CREATE TABLE Employees (ID INT PRIMARY KEY NOT NULL, NAME TEXT NOT NULL, AGE INT NOT NULL, ADDRESS CHAR(50), SALARY REAL);

**Inserting Data**

Use the INSERT INTO SQL statement to add new records to a table.
Specify the values to be inserted into each column of the table.
Example: INSERT INTO Employees (ID, NAME, AGE, ADDRESS, SALARY) VALUES (1, 'John', 30, 'New York', 50000.0);

**Querying Data**

Use the SELECT SQL statement to retrieve data from one or more tables.
You can specify conditions using the WHERE clause to filter the results.
Example: SELECT * FROM Employees WHERE AGE > 25;

**Updating Data**

Use the UPDATE SQL statement to modify existing records in a table.
Specify the columns to be updated along with the new values, and use the WHERE clause to specify the condition for the update.
Example: UPDATE Employees SET SALARY = 60000.0 WHERE NAME = 'John';

**Deleting Data**

Use the DELETE FROM SQL statement to remove records from a table.
Specify the conditions using the WHERE clause to determine which records to delete.
Example: DELETE FROM Employees WHERE AGE < 25;

**Transactions**

SQLite supports transactions, allowing you to group multiple SQL statements into a single atomic unit of work.
Use the BEGIN TRANSACTION, COMMIT, and ROLLBACK SQL statements to control transactions.

**Closing the Connection**

Always close the database connection using the close() method of the connection object after you finish working with the database.
This releases the database file and ensures that any changes are persisted to disk.

**Rollback**

rollback() method is used to rolls back any changes to the database since the last call to commit.

```python
try:
    # Connect to the database
    conn = sqlite3.connect('javatpoint.db')

    # Begin a transaction explicitly
    conn.execute("BEGIN TRANSACTION")

    # Execute the update query to increase the salary of employees with an age greater than 30
    conn.execute("""
        UPDATE Employees
        SET SALARY = SALARY * 1.1  -- Increase the salary by 10%
        WHERE AGE > 30
    """)

    # Roll back the transaction
    print("Rolling back changes...")
    conn.rollback()

except Exception as e:
    # If an exception occurs, print the error message
    print("An error occurred:", str(e))

else:
    # If no exception occurs, commit the changes to the database
    conn.commit()
    print("Changes committed successfully")

finally:
    # Close the connection
    conn.close()
```

```python
import sqlite3

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('iris.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Define the SQL query to check if the table exists
check_table_query = "SELECT name FROM sqlite_master WHERE type='table' AND name='Iris'"

# Execute the SQL query to check if the table exists
cursor.execute(check_table_query)

# Fetch the result
result = cursor.fetchone()

# If the table doesn't exist, create it
if not result:
    # Define the SQL query to create the table
    create_table_query = '''
    CREATE TABLE Iris (
        sepal_length REAL,
        sepal_width REAL,
        petal_length REAL,
        petal_width REAL,
        species TEXT
    )
    '''

    # Execute the SQL query to create the table
    cursor.execute(create_table_query)

    # Commit the transaction (save changes)
    conn.commit()

    print("Table 'Iris' created successfully.")
else:
    print("Table 'Iris' already exists.")

# Close the connection
conn.close()
```

```python
import sqlite3
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Define ridge loss function
def ridge_loss(B, B0, X, y, lamb):
    return np.sum((y - B0 - np.dot(X, B))**2) + lamb * np.sum(B**2)

def dRidgeLoss_dB0(B0, B, X, y):
    # Compute the dot product of X and B
    X_dot_B = np.dot(X, B)
    # Compute the residual term (y - B0 - X.dot(B))
    residual = y - B0 - X_dot_B
    # Compute the derivative of the loss function with respect to B0
    dL_dB0 = -2 * np.sum(residual)
    return dL_dB0

def dRidgeLoss_dB(B0, B, X, y, lamb):
    # Compute the residual term (y - B0 - X.dot(B))
    residual = y - B0 - np.dot(X, B)
    # Compute the derivative of the loss function with respect to B
    dL_dB = -2 * np.dot(X.T, residual) + 2 * lamb * B
    return dL_dB

# Define custom gradient descent algorithm
def gradient_descent(X, y, eta, lamb, epochs):
    # Initialize coefficients randomly
    np.random.seed(42)  # for reproducibility
    B0 = np.random.randn()
    B = np.random.randn(X.shape[1])

    # Loss storage
    lossGD = []


    for i in range(epochs):
        # Compute gradients
        dB0 = dRidgeLoss_dB0(B0, B, X, y)
        dB = dRidgeLoss_dB(B0, B, X, y, lamb)

        # Update coefficients
        B0 -= eta * dB0
        B -= eta * dB

        # Compute loss
        loss = ridge_loss(B, B0, X, y, lamb)
        lossGD.append(loss)

        # Print loss at each iteration
        print("Iteration:", i+1, "Loss:", loss)

    return B0, B, lossGD

# Connect to the SQLite database
conn = sqlite3.connect('iris.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute a query to select all data from the "Iris" table
cursor.execute("SELECT * FROM Iris")

# Fetch all rows from the result set
all_data = cursor.fetchall()

# Close the connection
conn.close()

# Convert fetched data to numpy array
data = np.array(all_data)

# Split the data into features (X) and target variable (y)
X = data[:, :-1]  # Features (all columns except the last one)
y_str = data[:, -1]  # Target variable (last column) as strings

# Encode the target variable using label encoding
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_str)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Set hyperparameters
eta = 0.001
lamb = 0.001
epochs = 1000

# Run gradient descents
B0, B, lossGD = gradient_descent(X_train_scaled, y_train, eta, lamb, epochs)
```
