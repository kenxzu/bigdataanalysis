```python
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("Apple.jpg")  # Assuming the image is in the same directory as your notebook

# OpenCV loads images in BGR format, so we need to convert it to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python

# Load the image
image = cv2.imread("Apple.jpg")

# Convert BGR to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using matplotlib
plt.imshow(image_gray, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# Simpan gambar
cv2.imwrite("apple-gray.png", image_gray)
```

```python
import csv

# Data to be written to the CSV file
data = [
    ['Name', 'Age', 'Country'],
    ['John', 30, 'USA'],
    ['Alice', 25, 'Canada'],
    ['Bob', 35, 'UK']
]

# Specify the file name
filename = 'example.csv'

# Open the file in write mode
csvfile = open(filename, 'w', newline='')

# Create a CSV writer object
csvwriter = csv.writer(csvfile)

# Write data to the CSV file
csvwriter.writerows(data)

# Close the file
csvfile.close()

print(f"CSV file '{filename}' has been written successfully.")
```

```python
import csv

# Data to be written to the CSV file

fields = ['Name', 'Branch', 'Year']

rows = [
    ['John', '30', 'USA'],
    ['Alice', '25', 'Canada'],
    ['Bob', '35', 'UK']
]

# Specify the file name
filename = 'example2.csv'

# Writing to CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Writing data to the CSV file
    csvwriter.writerows(fields)
    csvwriter.writerows(rows)

print(f"CSV file '{filename}' has been written successfully.")
```

```python
# Membaca isi file csv
import csv

# name of csv file
filename = "example.csv"

# reading from csv file
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)

    # extracting field names (header)
    fields = next(csvreader)

    # printing the field names
    print(f'Field names: {", ".join(fields)}')

    # printing each row
    for row in csvreader:
        print(', '.join(row))
```
